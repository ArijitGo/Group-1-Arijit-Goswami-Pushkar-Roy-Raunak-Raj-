```python
# ==========================================================
# IMPORTS
# ==========================================================
import numpy as np
import time
import matplotlib.pyplot as plt

# ==========================================================
# CONSTANT PARAMETERS
# ==========================================================
POP_SIZE = 100
ITERATIONS = 300
TEACHING_FACTOR = 2

# ==========================================================
# INITIAL POPULATION (random integer assignments)
# ==========================================================
def generate_initial_population(pop_size, m, n):
    return np.random.randint(0, m, size=(pop_size, n))

# ==========================================================
# REPAIR STUDENT (random move to a feasible agent)
# ==========================================================
def repair_student_random(student, C, R, B):
    n = len(C[0])   # number of Jobs
    m = len(C)      # number of Agents

    agents = student.astype(int).copy()
    resource_used = np.zeros(m, dtype=int)

    for j, a in enumerate(agents):
        resource_used[a] += R[a][j]

    for a in range(m):
        while resource_used[a] > B[a]:
            jobs = [j for j in range(n) if agents[j] == a]
            if not jobs:
                break
            j = np.random.choice(jobs)
            feasible_agents = []
            for b in range(m):
                if b != a and resource_used[b] + R[b][j] <= B[b]:
                    feasible_agents.append(b)
            if feasible_agents:
                new_agent = np.random.choice(feasible_agents)
                resource_used[a] -= R[a][j]
                resource_used[new_agent] += R[new_agent][j]
                agents[j] = new_agent
            else:
                break
    return agents

# ==========================================================
# FITNESS FUNCTION (maximisation: total profit)
# ==========================================================
def fitness(student, C, R, B):
    total_cost = 0
    for j, a in enumerate(student):
        total_cost += C[a][j]
    return total_cost

# ==========================================================
# FEASIBILITY CHECK
# ==========================================================
def is_feasible(student, R, B):
    m = len(R)
    used = np.zeros(m, dtype=int)
    for j, a in enumerate(student):
        used[a] += R[a][j]
        if used[a] > B[a]:
            return False
    return True

# ==========================================================
# TEACHING LEARNING BASED OPTIMIZATION (with repair)
# ==========================================================
def teaching_learning_based_optimization(C, R, B):
    m = len(C)      # agents
    n = len(C[0])   # jobs

    # Initial random population (may be infeasible)
    population = generate_initial_population(POP_SIZE, m, n)
    fitness_values = np.array([fitness(population[i], C, R, B) for i in range(POP_SIZE)])

    # Global best (maximisation)
    best_idx = np.argmax(fitness_values)
    best_solution = population[best_idx].copy()
    best_fitness = fitness_values[best_idx]

    best_fitness_per_gen = []

    for gen in range(ITERATIONS):
        for i in range(POP_SIZE):
            # ---------- Teaching Phase ----------
            r1 = np.random.rand()
            x_best_idx = np.argmax(fitness_values)
            x_best = population[x_best_idx].copy()
            x_mean = np.mean(population, axis=0)
            x_new = population[i] + r1 * (x_best - TEACHING_FACTOR * x_mean)
            x_new = np.clip(np.round(x_new), 0, m - 1).astype(int)

            # Accept if better (no repair needed in teaching phase)
            f_new = fitness(x_new, C, R, B)
            if f_new > fitness_values[i]:
                population[i] = x_new
                fitness_values[i] = f_new
                if f_new > best_fitness:
                    best_fitness = f_new
                    best_solution = x_new.copy()

            # ---------- Learner Phase ----------
            # Pick random partner
            indices = list(range(POP_SIZE))
            indices.remove(i)
            p_idx = np.random.choice(indices)
            x_p = population[p_idx]
            f_p = fitness_values[p_idx]

            r2 = np.random.rand()
            if f_p > fitness_values[i]:
                x_new = population[i] + r2 * (population[i] - x_p)
            else:
                x_new = population[i] - r2 * (population[i] - x_p)
            x_new = np.clip(np.round(x_new), 0, m - 1).astype(int)

            # Repair if infeasible
            if not is_feasible(x_new, R, B):
                x_new = repair_student_random(x_new, C, R, B)

            f_new = fitness(x_new, C, R, B)
            if f_new > fitness_values[i]:
                population[i] = x_new
                fitness_values[i] = f_new
                if f_new > best_fitness:
                    best_fitness = f_new
                    best_solution = x_new.copy()

        best_fitness_per_gen.append(best_fitness)

    # Final repair of the best solution (to guarantee feasibility)
    if not is_feasible(best_solution, R, B):
        best_solution = repair_student_random(best_solution, C, R, B)
    final_profit = fitness(best_solution, C, R, B)
    return best_solution, final_profit, best_fitness_per_gen

# ==========================================================
# RUN 20 TIMES AND COLLECT STATISTICS
# ==========================================================
num_runs = 20
results = []          # (profit, time, history)
best_overall = (-1e9, None, None)

print("Running TLBO 20 times (repair‑based feasibility)...")
for run in range(num_runs):
    start_time = time.perf_counter()
    assign, profit, hist = teaching_learning_based_optimization(cost, resource, capacity)
    elapsed = time.perf_counter() - start_time
    results.append((profit, elapsed, hist))
    print(f"Run {run+1:2d}: Profit = {profit:6.2f}, Time = {elapsed:.2f} s")
    if profit > best_overall[0]:
        best_overall = (profit, assign, hist)

# Statistics
profits = [r[0] for r in results]
times = [r[1] for r in results]
avg_profit = np.mean(profits)
std_profit = np.std(profits)
best_profit = np.max(profits)
worst_profit = np.min(profits)
avg_time = np.mean(times)
total_time = np.sum(times)

print("\n=== Summary over {} runs ===".format(num_runs))
print(f"Average profit    : {avg_profit:.2f} ± {std_profit:.2f}")
print(f"Best profit       : {best_profit:.2f}")
print(f"Worst profit      : {worst_profit:.2f}")
print(f"Average time      : {avg_time:.2f} s")
print(f"Total time        : {total_time:.2f} s")

# ==========================================================
# CONVERGENCE PLOT (all runs + average, NO shaded region)
# ==========================================================
plt.figure(figsize=(10, 6))
max_len = max(len(h) for _,_,h in results)
padded = []
for _,_,h in results:
    pad = [h[-1]] * (max_len - len(h))
    padded.append(np.concatenate([h, pad]))
hist_array = np.array(padded)
mean_hist = np.mean(hist_array, axis=0)
iters = range(max_len)

# Plot each run (semi-transparent blue)
for i in range(num_runs):
    plt.plot(iters, hist_array[i], color='blue', alpha=0.2, linewidth=0.8)

# Plot average profit (solid red)
plt.plot(iters, mean_hist, color='red', linewidth=2, label='Average profit')

# No fill_between (shaded region removed)

plt.title('TLBO Convergence – Profit over Generations (20 runs) – Repair‑based')
plt.xlabel('Generation')
plt.ylabel('Total Profit')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig('convergence_profit.png')
plt.show()
```
