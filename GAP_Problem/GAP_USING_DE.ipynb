```python
import numpy as np
import random
import time
import matplotlib.pyplot as plt

# ----------------------------------------------------------------------
# Parameters (UNCHANGED)
# ----------------------------------------------------------------------
m = len(cost)          # number of agents (10)
n = len(cost[0])       # number of jobs (60)
Np = 100               # population size
T = 300                # number of generations
F = 0.8                # scaling factor
pc = 0.8               # crossover probability
lambda_penalty = 1e9   # very high penalty (repair ensures feasibility)

# ----------------------------------------------------------------------
# Helper: compute resource usage per agent
# ----------------------------------------------------------------------
def get_usage(x):
    used = [0] * m
    for j, agent in enumerate(x):
        agent_idx = agent - 1
        used[agent_idx] += resource[agent_idx][j]
    return used

# ----------------------------------------------------------------------
# Penalty computation
# ----------------------------------------------------------------------
def compute_penalty(x):
    used = get_usage(x)
    excess = 0
    for i in range(m):
        if used[i] > capacity[i]:
            excess += (used[i] - capacity[i])
    return lambda_penalty * excess

# ----------------------------------------------------------------------
# Repair function: makes a solution feasible by reassigning jobs
# ----------------------------------------------------------------------
def repair(x):
    """
    Try to repair an infeasible assignment.
    Returns a feasible assignment (may not be optimal).
    """
    x = list(x)          # copy
    used = get_usage(x)

    # Already feasible?
    if all(used[i] <= capacity[i] for i in range(m)):
        return x

    overloaded = [i for i in range(m) if used[i] > capacity[i]]
    max_iter = 100
    for _ in range(max_iter):
        if not overloaded:
            break

        # Most overloaded agent
        i = max(overloaded, key=lambda i: used[i] - capacity[i])
        jobs_assigned = [j for j, a in enumerate(x) if a == i+1]
        if not jobs_assigned:
            overloaded.remove(i)
            continue

        # Select job with largest resource consumption from this agent
        job = max(jobs_assigned, key=lambda j: resource[i][j])

        # Find an agent with enough free capacity to take this job
        best_target = None
        best_cap = -1
        for k in range(m):
            if k == i:
                continue
            if used[k] + resource[i][job] <= capacity[k]:
                if capacity[k] - used[k] > best_cap:
                    best_cap = capacity[k] - used[k]
                    best_target = k

        if best_target is not None:
            # Move job
            x[job] = best_target + 1
            used[i] -= resource[i][job]
            used[best_target] += resource[best_target][job]
        else:
            # No target found; remove this job from consideration
            jobs_assigned.remove(job)
            if not jobs_assigned:
                overloaded.remove(i)

        overloaded = [i for i in range(m) if used[i] > capacity[i]]

    # Final check – if still infeasible, return the original (will be heavily penalized)
    if all(used[i] <= capacity[i] for i in range(m)):
        return x
    else:
        return x

# ----------------------------------------------------------------------
# Fitness function (penalty only for residual infeasibility)
# ----------------------------------------------------------------------
def fitness(x):
    profit = 0.0
    used = [0] * m
    for j, agent in enumerate(x):
        agent_idx = agent - 1
        profit += cost[agent_idx][j]
        used[agent_idx] += resource[agent_idx][j]
    penalty = 0.0
    for i in range(m):
        if used[i] > capacity[i]:
            penalty += (used[i] - capacity[i])
    return profit - lambda_penalty * penalty

# ----------------------------------------------------------------------
# Differential Evolution with repair and monotonic profit tracking
# ----------------------------------------------------------------------
def de(seed=None):
    if seed is not None:
        random.seed(seed)

    # Initialise population with feasible solutions
    pop = []
    for _ in range(Np):
        x = [random.randint(1, m) for _ in range(n)]
        x = repair(x)
        pop.append(x)
    fit = [fitness(x) for x in pop]

    # Best according to penalized fitness
    best_idx = np.argmax(fit)
    best_x = pop[best_idx][:]

    # Track best actual profit seen so far (monotonic)
    best_actual_profit = sum(cost[best_x[j]-1][j] for j in range(n))
    history = [best_actual_profit]   # record starting best

    for gen in range(T):
        trial = []
        for i in range(Np):
            # Mutation (DE/rand/1)
            indices = list(range(Np))
            indices.remove(i)
            r1, r2, r3 = random.sample(indices, 3)
            v = []
            for j in range(n):
                val = pop[r1][j] + F * (pop[r2][j] - pop[r3][j])
                vj = int(round(val))
                if vj < 1:
                    vj = 1
                if vj > m:
                    vj = m
                v.append(vj)

            # Crossover
            delta = random.randint(0, n-1)
            u = []
            for j in range(n):
                if random.random() <= pc or j == delta:
                    u.append(v[j])
                else:
                    u.append(pop[i][j])
            u = repair(u)
            trial.append(u)

        # Selection
        for i in range(Np):
            fit_u = fitness(trial[i])
            if fit_u >= fit[i]:
                pop[i] = trial[i]
                fit[i] = fit_u
                # Update best_x if penalized fitness improved
                if fit_u > fitness(best_x):
                    best_x = trial[i][:]

        # Compute actual profit of current best_x (may be lower than best_actual_profit)
        current_actual = sum(cost[best_x[j]-1][j] for j in range(n))
        # Update best actual profit seen so far (monotonic)
        if current_actual > best_actual_profit:
            best_actual_profit = current_actual
        history.append(best_actual_profit)

    # Final profit is the best actual profit found (monotonic)
    final_profit = best_actual_profit
    final_penalty = compute_penalty(best_x)   # may not be zero if best_x is infeasible
    return final_profit, final_penalty, history

# ----------------------------------------------------------------------
# Run the algorithm 20 times
# ----------------------------------------------------------------------
num_runs = 20
profits = []
penalties = []
times = []
convergence_histories = []

for run in range(num_runs):
    start = time.perf_counter()
    profit, penalty, hist = de(seed=run)
    end = time.perf_counter()
    profits.append(profit)
    penalties.append(penalty)
    times.append(end - start)
    convergence_histories.append(hist)
    print(f"Run {run+1}: Profit = {profit:.2f}, Penalty = {penalty:.2f}, Time = {end-start:.2f} s")

# Summary statistics
avg_profit = np.mean(profits)
std_profit = np.std(profits)
best_profit = max(profits)
worst_profit = min(profits)
avg_time = np.mean(times)
total_time = np.sum(times)

print("\n=== Summary over {} runs ===".format(num_runs))
print(f"Average profit    : {avg_profit:.2f} ± {std_profit:.2f}")
print(f"Best profit       : {best_profit:.2f}")
print(f"Worst profit      : {worst_profit:.2f}")
print(f"Average time      : {avg_time:.2f} s")
print(f"Total time        : {total_time:.2f} s")
print(f"All final penalties zero? {all(p == 0 for p in penalties)}")

# ----------------------------------------------------------------------
# Convergence plot: all 20 runs + average (monotonic)
# ----------------------------------------------------------------------
plt.figure(figsize=(12, 7))

# Plot each run with low opacity (thin lines)
for i, hist in enumerate(convergence_histories):
    plt.plot(range(1, T+2), hist, 'b-', linewidth=0.5, alpha=0.3)   # T+1 points

# Plot the average curve (thick, high contrast)
avg_history = np.mean(convergence_histories, axis=0)
plt.plot(range(1, T+2), avg_history, 'r-', linewidth=3, label='Average over 20 runs')

plt.xlabel('Generation')
plt.ylabel('Best Profit (monotonically increasing)')
plt.title('Convergence of Differential Evolution (monotonic profit)')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```
