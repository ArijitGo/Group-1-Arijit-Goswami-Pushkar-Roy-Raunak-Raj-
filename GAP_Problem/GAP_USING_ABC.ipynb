```python
import random
import numpy as np
import time
import matplotlib.pyplot as plt

NUM_AGENTS = 10
NUM_JOBS = 60
LAMBDA = 1_000_000   # penalty coefficient

def evaluate(solution):
    """Return profit - LAMBDA * total resource violation."""
    profit = 0
    used = [0] * NUM_AGENTS
    for job, agent in enumerate(solution):
        profit += cost[agent][job]
        used[agent] += resource[agent][job]
    violation = sum(max(0, used[i] - capacity[i]) for i in range(NUM_AGENTS))
    return profit - LAMBDA * violation

def true_profit_and_feasibility(solution):
    """Return (profit, feasible_flag) without penalty."""
    profit = 0
    used = [0] * NUM_AGENTS
    for job, agent in enumerate(solution):
        profit += cost[agent][job]
        used[agent] += resource[agent][job]
    feasible = all(used[i] <= capacity[i] for i in range(NUM_AGENTS))
    return profit, feasible

def compute_penalty(solution):
    """Return total penalty = LAMBDA * total resource violation."""
    used = [0] * NUM_AGENTS
    for job, agent in enumerate(solution):
        used[agent] += resource[agent][job]
    violation = sum(max(0, used[i] - capacity[i]) for i in range(NUM_AGENTS))
    return LAMBDA * violation

def generate_neighbour(source, partner):
    """Discrete neighbour: copy partner or random agent for one job."""
    new_sol = source.copy()
    job = random.randint(0, NUM_JOBS - 1)
    if random.random() < 0.5:
        new_sol[job] = partner[job]
    else:
        new_sol[job] = random.randint(0, NUM_AGENTS - 1)
    return new_sol

def abc_gap(Np, T, limit, seed=None):
    """Run ABC and return (best_solution, best_true_profit, convergence_history)."""
    if seed is not None:
        random.seed(seed)
        np.random.seed(seed)

    # Initialisation
    population = []
    fitness = []
    trial = []
    for _ in range(Np):
        sol = [random.randint(0, NUM_AGENTS - 1) for _ in range(NUM_JOBS)]
        fit = evaluate(sol)
        population.append(sol)
        fitness.append(fit)
        trial.append(0)

    # Best tracking
    best_idx = np.argmax(fitness)
    best_sol = population[best_idx].copy()
    best_true_profit, _ = true_profit_and_feasibility(best_sol)
    history = []   # best true profit per generation

    for gen in range(T):
        # Employed Bee Phase
        for i in range(Np):
            partner = random.choice([p for p in range(Np) if p != i])
            new_sol = generate_neighbour(population[i], population[partner])
            new_fit = evaluate(new_sol)
            if new_fit > fitness[i]:
                population[i] = new_sol
                fitness[i] = new_fit
                trial[i] = 0
            else:
                trial[i] += 1

        # Probability calculation (shift if necessary)
        total_fit = sum(fitness)
        if total_fit <= 0:
            shift = -total_fit + 1e-6
            prob = [(f + shift) / (total_fit + shift * Np) for f in fitness]
        else:
            prob = [f / total_fit for f in fitness]

        # Onlooker Bee Phase
        m = 0
        n = 0
        while m < Np:
            if random.random() < prob[n]:
                partner = random.choice([p for p in range(Np) if p != n])
                new_sol = generate_neighbour(population[n], population[partner])
                new_fit = evaluate(new_sol)
                if new_fit > fitness[n]:
                    population[n] = new_sol
                    fitness[n] = new_fit
                    trial[n] = 0
                else:
                    trial[n] += 1
                m += 1
            n = (n + 1) % Np

        # Scout Bee Phase
        for i in range(Np):
            if trial[i] > limit:
                new_sol = [random.randint(0, NUM_AGENTS - 1) for _ in range(NUM_JOBS)]
                new_fit = evaluate(new_sol)
                population[i] = new_sol
                fitness[i] = new_fit
                trial[i] = 0

        # Update best solution
        current_best_idx = np.argmax(fitness)
        current_best_sol = population[current_best_idx]
        current_true_profit, _ = true_profit_and_feasibility(current_best_sol)
        if current_true_profit > best_true_profit:
            best_true_profit = current_true_profit
            best_sol = current_best_sol.copy()
        history.append(best_true_profit)

    return best_sol, best_true_profit, history

# ------------------------------------------------------------
# Main: run 20 times, collect statistics, plot convergence
# ------------------------------------------------------------

def main():
    NUM_RUNS = 20
    results = []          # (profit, penalty, feasible)
    all_histories = []    # store history of each run
    best_profit_overall = -float('inf')
    best_solution_overall = None
    best_run_history = None

    print("Running ABC 20 times...\n")
    for run in range(1, NUM_RUNS + 1):
        start_time = time.time()
        sol, profit, history = abc_gap(Np=100, T=300, limit=30, seed=run)
        elapsed = time.time() - start_time
        _, feasible = true_profit_and_feasibility(sol)
        penalty = compute_penalty(sol)
        results.append((profit, penalty, feasible))
        all_histories.append(history)
        if profit > best_profit_overall:
            best_profit_overall = profit
            best_solution_overall = sol
            best_run_history = history
        print(f"Run {run:2d}: Profit = {profit:.2f}, Penalty = {penalty:.2f}, Time = {elapsed:.2f} s, Feasible = {feasible}")

    # Summary statistics
    profits = [r[0] for r in results]
    penalties = [r[1] for r in results]
    feasibles = [r[2] for r in results]
    feasible_count = sum(feasibles)
    avg_profit = np.mean(profits)
    std_profit = np.std(profits)
    best_profit = max(profits)
    worst_profit = min(profits)
    avg_time = np.mean([r[2] for r in results])  # time is not in results, we didn't store it; we'll recompute from earlier loop

def main():
    NUM_RUNS = 20
    results = []          # (profit, penalty, feasible)
    times = []
    all_histories = []

    print("Running ABC 20 times...\n")
    for run in range(1, NUM_RUNS + 1):
        start_time = time.time()
        sol, profit, history = abc_gap(Np=200, T=300, limit=30, seed=run)
        elapsed = time.time() - start_time
        times.append(elapsed)
        _, feasible = true_profit_and_feasibility(sol)
        penalty = compute_penalty(sol)
        results.append((profit, penalty, feasible))
        all_histories.append(history)
        print(f"Run {run:2d}: Profit = {profit:.2f}, Penalty = {penalty:.2f}, Time = {elapsed:.2f} s, Feasible = {feasible}")

    profits = [r[0] for r in results]
    penalties = [r[1] for r in results]
    feasibles = [r[2] for r in results]
    avg_profit = np.mean(profits)
    std_profit = np.std(profits)
    best_profit = max(profits)
    worst_profit = min(profits)
    avg_time = np.mean(times)
    total_time = sum(times)

    print("\n=== Summary over {} runs ===".format(NUM_RUNS))
    print(f"Average profit    : {avg_profit:.2f} ± {std_profit:.2f}")
    print(f"Best profit       : {best_profit:.2f}")
    print(f"Worst profit      : {worst_profit:.2f}")
    print(f"Average time      : {avg_time:.2f} s")
    print(f"Total time        : {total_time:.2f} s")
    print(f"Feasible runs     : {sum(feasibles)}/{NUM_RUNS}")
    print(f"All final penalties zero? {all(p == 0 for p in penalties)}")

    # Plot convergence
    plt.figure(figsize=(12, 6))
    generations = range(300)
    for hist in all_histories:
        plt.plot(generations, hist, color='blue', alpha=0.2, linewidth=0.8)
    avg_history = np.mean(all_histories, axis=0)
    plt.plot(generations, avg_history, color='red', linewidth=2.5, label='Average (20 runs)')
    plt.xlabel('Generation', fontsize=12)
    plt.ylabel('Best True Profit', fontsize=12)
    plt.title('Convergence of ABC over 20 Runs', fontsize=14)
    plt.grid(True, alpha=0.3)
    plt.legend()
    plt.tight_layout()
    plt.savefig('abc_all_runs_convergence.png', dpi=150)
    plt.show()

if __name__ == "__main__":
    main()
```
