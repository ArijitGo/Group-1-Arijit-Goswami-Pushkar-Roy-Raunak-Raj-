```python
import time
import numpy as np
import matplotlib.pyplot as plt

# Problem dimensions
cost = np.array(cost)
resource = np.array(resource)
capacity = np.array(capacity)
m, n = cost.shape

# ======================================================
# RCGA parameters (UNCHANGED)
# ======================================================
POP_SIZE = 200
GENERATIONS = 300
ELITE_COUNT = 2
PENALTY = 1e6

LOW, HIGH = 0.0, float(m) - 1e-10
CROSSOVER_RATE = 0.8
ETA_C = 20
MUTATION_RATE = 0.1
ETA_M = 20

pop_size = POP_SIZE
n_jobs = n
crossover_rate = CROSSOVER_RATE
eta_c = ETA_C
eta_m = ETA_M
mutation_rate = MUTATION_RATE
Xl = LOW
Xu = HIGH

# ======================================================
# Genetic operators (EXACTLY as provided)
# ======================================================
def sbx_crossover(pop):
    new_pop = pop.copy()
    for i in range(0, pop_size, 2):
        if np.random.rand() < crossover_rate:
            p1 = pop[i]
            p2 = pop[i+1]
            u = np.random.rand(n_jobs)
            beta = np.where(
                u <= 0.5,
                (2*u)**(1/(eta_c+1)),
                (1/(2*(1-u)))**(1/(eta_c+1))
            )
            c1 = 0.5*((1+beta)*p1 + (1-beta)*p2)
            c2 = 0.5*((1-beta)*p1 + (1+beta)*p2)
            new_pop[i] = np.clip(c1, Xl, Xu)
            new_pop[i+1] = np.clip(c2, Xl, Xu)
    return new_pop

def polynomial_mutation(pop):
    mutation_mask = np.random.rand(*pop.shape) < mutation_rate
    r = np.random.rand(*pop.shape)
    delta = np.where(
        r < 0.5,
        (2*r)**(1/(eta_m+1)) - 1,
        1 - (2*(1-r))**(1/(eta_m+1))
    )
    pop[mutation_mask] += delta[mutation_mask] * (Xu - Xl)
    return np.clip(pop, Xl, Xu)

def decode(chromosome):
    return np.minimum(np.floor(chromosome).astype(int), m-1)

def compute_penalty(assignment):
    used = np.bincount(assignment, weights=resource[assignment, np.arange(n)], minlength=m)
    violations = np.maximum(0, used - capacity)
    return PENALTY * violations.sum()

def evaluate(chromosome):
    assign = decode(chromosome)
    profit = cost[assign, np.arange(n)].sum()
    used = np.bincount(assign, weights=resource[assign, np.arange(n)], minlength=m)
    violations = np.maximum(0, used - capacity)
    total_violation = violations.sum()
    fitness = profit - PENALTY * total_violation
    return fitness, profit, used, violations

def is_feasible(chromosome):
    assign = decode(chromosome)
    used = np.bincount(assign, weights=resource[assign, np.arange(n)], minlength=m)
    return np.all(used <= capacity)

def initialize_population():
    return np.random.uniform(LOW, HIGH, (POP_SIZE, n))

def tournament_selection(pop, fitness, k=3):
    idx = np.random.choice(POP_SIZE, k, replace=False)
    best_idx = idx[np.argmax(fitness[idx])]
    return pop[best_idx].copy()

# ======================================================
# Single GA run – now tracks best profit (any solution)
# ======================================================
def run_single_ga():
    pop = initialize_population()
    best_feasible = None
    best_feasible_profit = -np.inf
    best_infeasible = None
    best_infeasible_profit = -np.inf

    # <-- CHANGE: history will store the best profit (feasible or not) so far
    history = []   # best profit overall (any solution)

    for gen in range(GENERATIONS):
        fitness = np.array([evaluate(ind)[0] for ind in pop])
        profits = np.array([evaluate(ind)[1] for ind in pop])   # true profit

        # Track best feasible and best infeasible solutions (for final return)
        for i, ind in enumerate(pop):
            if is_feasible(ind):
                prof = profits[i]
                if prof > best_feasible_profit:
                    best_feasible_profit = prof
                    best_feasible = ind.copy()
            else:
                prof = profits[i]
                if prof > best_infeasible_profit:
                    best_infeasible_profit = prof
                    best_infeasible = ind.copy()

        # <-- CHANGE: record the best profit (any solution) in this generation
        gen_best_profit = np.max(profits)
        if gen_best_profit > (history[-1] if history else -np.inf):
            history.append(gen_best_profit)
        else:
            history.append(history[-1] if history else -np.inf)

        # Elitism (unchanged)
        elite_indices = np.argsort(fitness)[-ELITE_COUNT:]
        elites = pop[elite_indices].copy()

        mating_pool = np.array([tournament_selection(pop, fitness) for _ in range(POP_SIZE)])
        offspring = sbx_crossover(mating_pool)
        offspring = polynomial_mutation(offspring)

        off_fitness = np.array([evaluate(ind)[0] for ind in offspring])
        worst_idx = np.argsort(off_fitness)[:ELITE_COUNT]
        offspring[worst_idx] = elites
        pop = offspring

    # Return best feasible if exists, else best infeasible (unchanged)
    if best_feasible is not None:
        final_profit = evaluate(best_feasible)[1]
        final_penalty = compute_penalty(decode(best_feasible))
        return final_profit, final_penalty, True, history
    else:
        if best_infeasible is not None:
            final_profit = evaluate(best_infeasible)[1]
            final_penalty = compute_penalty(decode(best_infeasible))
            return final_profit, final_penalty, False, history
        else:
            return -np.inf, np.inf, False, history

# ======================================================
# Plot convergence – shows best profit for ALL runs
# ======================================================
def plot_convergence(all_histories):
    plt.figure(figsize=(10, 6))
    generations = np.arange(1, GENERATIONS + 1)

    # Plot each run's history (all runs, even infeasible)
    for i, hist in enumerate(all_histories):
        plt.plot(generations, hist, alpha=0.3, linewidth=0.8,
                 label=f'Run {i+1}' if i < 5 else "")   # label only first few

    # Average over all runs
    avg_history = np.mean(all_histories, axis=0)
    plt.plot(generations, avg_history, 'k-', linewidth=2.5, label='Average (all runs)')

    plt.xlabel('Generation')
    plt.ylabel('Best Profit (may be infeasible)')
    plt.title('Convergence of Real-Coded GA')
    plt.grid(alpha=0.3)
    plt.legend(loc='lower right')
    plt.tight_layout()
    plt.show()

# ======================================================
# Main: 20 runs
# ======================================================
def main():
    num_runs = 20
    results = []
    times = []
    all_histories = []

    print("Running GA 20 times...\n")
    for run in range(1, num_runs + 1):
        np.random.seed(run)
        start = time.time()
        profit, penalty, feasible, history = run_single_ga()
        elapsed = time.time() - start
        results.append((profit, penalty, feasible))
        times.append(elapsed)
        all_histories.append(history)
        print(f"Run {run:2d}: Profit = {profit:.2f}, Penalty = {penalty:.2f}, Time = {elapsed:.2f}s, Feasible = {feasible}")

    profits = [r[0] for r in results]
    penalties = [r[1] for r in results]
    feasibles = [r[2] for r in results]

    print("\n" + "="*50)
    print("=== Summary over 20 runs ===")
    print(f"Average profit    : {np.mean(profits):.2f} ± {np.std(profits):.2f}")
    print(f"Best profit       : {np.max(profits):.2f}")
    print(f"Worst profit      : {np.min(profits):.2f}")
    print(f"Average time      : {np.mean(times):.2f} s")
    print(f"Total time        : {np.sum(times):.2f} s")
    print(f"All final penalties zero? {all(p == 0 for p in penalties)}")

    plot_convergence(all_histories)

if __name__ == "__main__":
    main()
```
