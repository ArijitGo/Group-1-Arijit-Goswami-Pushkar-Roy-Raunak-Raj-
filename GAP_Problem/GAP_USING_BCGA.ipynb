```python
import random
import time
import numpy as np
import matplotlib.pyplot as plt
from heapq import heappush, heappop

try:
    cost_np = np.array(cost)
    resource_np = np.array(resource)
    capacity_np = np.array(capacity)
except NameError:
    raise NameError("Please define cost, resource, and capacity matrices before running this code.")

NUM_AGENTS = cost_np.shape[0]
NUM_JOBS = cost_np.shape[1]

assert resource_np.shape == (NUM_AGENTS, NUM_JOBS), "Resource matrix shape mismatch"
assert len(capacity_np) == NUM_AGENTS, "Capacity length mismatch"
print(f"Data loaded: {NUM_AGENTS} agents, {NUM_JOBS} jobs.")

# ============================
# GA Parameters
# ============================
POP_SIZE = 100
GENERATIONS = 300
CROSSOVER_RATE = 0.8
ELITE_SIZE = 2
CHROM_LENGTH = NUM_JOBS * 4
MUTATION_RATE = 0.1

# ============================
# Robust Repair Operator
# ============================
def is_feasible(assignment):
    usage = np.zeros(NUM_AGENTS, dtype=int)
    for j, a in enumerate(assignment):
        usage[a] += resource_np[a, j]
    return np.all(usage <= capacity_np)

def compute_usage(assignment):
    usage = np.zeros(NUM_AGENTS, dtype=int)
    for j, a in enumerate(assignment):
        usage[a] += resource_np[a, j]
    return usage

def repair_assignment(assignment):
    """
    Make assignment feasible by moving jobs from overloaded agents to underloaded ones.
    Returns a feasible assignment if one exists, otherwise raises ValueError.
    """
    assign = assignment[:]
    usage = compute_usage(assign)
    
    # Pre-check: if for any agent, total possible demand (all jobs assigned to it) < capacity, but that's not the issue.
    # Main loop: while any agent overloaded
    max_iter = 1000
    for _ in range(max_iter):
        if np.all(usage <= capacity_np):
            return assign
        
        # Find overloaded agents (those with usage > capacity)
        overloaded = [i for i in range(NUM_AGENTS) if usage[i] > capacity_np[i]]
        if not overloaded:
            break
        
        # Choose the most overloaded agent
        agent = max(overloaded, key=lambda i: usage[i] - capacity_np[i])
        excess = usage[agent] - capacity_np[agent]
        
        # Find jobs assigned to this agent
        jobs = [j for j, a in enumerate(assign) if a == agent]
        if not jobs:
            break
        
        # Try to move a job that reduces overload the most (largest resource consumption)
        # Sort jobs by resource consumption descending
        jobs.sort(key=lambda j: resource_np[agent, j], reverse=True)
        
        moved = False
        for j in jobs:
            # Find target agents with enough spare capacity to take this job
            spare = [i for i in range(NUM_AGENTS) if i != agent and usage[i] + resource_np[i, j] <= capacity_np[i]]
            if spare:
                # Choose the agent with the largest spare capacity after adding the job
                target = max(spare, key=lambda i: capacity_np[i] - (usage[i] + resource_np[i, j]))
                # Move job
                usage[agent] -= resource_np[agent, j]
                usage[target] += resource_np[target, j]
                assign[j] = target
                moved = True
                break
        
        if not moved:
            # No direct move works – try swapping jobs between overloaded and underloaded agents
            # Find underloaded agents (usage < capacity)
            underloaded = [i for i in range(NUM_AGENTS) if usage[i] < capacity_np[i]]
            if underloaded:
                # For each job on overloaded agent, try to swap with a job on an underloaded agent
                for j in jobs:
                    for u in underloaded:
                        # Find a job on u that we could swap with j
                        jobs_u = [jj for jj, a in enumerate(assign) if a == u]
                        for jj in jobs_u:
                            # Check if swapping j and jj keeps both agents within capacity
                            # New usage if we move j to u and jj to agent
                            new_usage_u = usage[u] - resource_np[u, jj] + resource_np[u, j]
                            new_usage_agent = usage[agent] - resource_np[agent, j] + resource_np[agent, jj]
                            if new_usage_u <= capacity_np[u] and new_usage_agent <= capacity_np[agent]:
                                # Perform swap
                                usage[agent] = new_usage_agent
                                usage[u] = new_usage_u
                                assign[j], assign[jj] = assign[jj], assign[j]
                                moved = True
                                break
                        if moved:
                            break
                    if moved:
                        break
            if not moved:
                # Cannot repair – problem is infeasible
                raise ValueError("Infeasible problem: cannot satisfy capacity constraints even after moving/swapping jobs.")
    
    if not np.all(usage <= capacity_np):
        raise ValueError("Repair failed – problem likely infeasible.")
    return assign

def decode_chromosome(chromosome):
    """Decode binary chromosome to agent assignment, then repair to feasibility."""
    raw_assignment = []
    for j in range(NUM_JOBS):
        start = j * 4
        bits = chromosome[start:start+4]
        val = bits[0]*8 + bits[1]*4 + bits[2]*2 + bits[3]
        agent = val % NUM_AGENTS
        raw_assignment.append(agent)
    # Repair to satisfy capacity constraints
    return repair_assignment(raw_assignment)

def compute_profit(assignment):
    return sum(cost_np[agent, j] for j, agent in enumerate(assignment))

def random_chromosome():
    return [random.randint(0, 1) for _ in range(CHROM_LENGTH)]

def initialize_population():
    return [random_chromosome() for _ in range(POP_SIZE)]

def fitness(chromosome):
    assignment = decode_chromosome(chromosome)  # already feasible
    return compute_profit(assignment)

def selection(population, fitnesses, k=3):
    indices = random.sample(range(len(population)), k)
    best_idx = indices[0]
    best_fit = fitnesses[best_idx]
    for idx in indices[1:]:
        if fitnesses[idx] > best_fit:
            best_fit = fitnesses[idx]
            best_idx = idx
    return population[best_idx][:]

def crossover(parent1, parent2):
    if random.random() < CROSSOVER_RATE:
        point = random.randint(1, CHROM_LENGTH - 1)
        child1 = parent1[:point] + parent2[point:]
        child2 = parent2[:point] + parent1[point:]
    else:
        child1 = parent1[:]
        child2 = parent2[:]
    return child1, child2

def mutate(chromosome):
    for i in range(CHROM_LENGTH):
        if random.random() < MUTATION_RATE:
            chromosome[i] = 1 - chromosome[i]
    return chromosome

def genetic_algorithm():
    population = initialize_population()
    best_profit_per_gen = []
    best_chromosome = None
    best_profit = -float('inf')

    for gen in range(GENERATIONS):
        fitnesses = []
        for ind in population:
            try:
                fit = fitness(ind)
            except ValueError as e:
                # If repair fails, assign very low fitness to this individual
                fit = -1e9
            fitnesses.append(fit)

        gen_best_profit = max(fitnesses)
        best_profit_per_gen.append(gen_best_profit)

        if gen_best_profit > best_profit:
            best_profit = gen_best_profit
            best_idx = fitnesses.index(gen_best_profit)
            best_chromosome = population[best_idx][:]

        # Selection
        new_population = []
        sorted_indices = sorted(range(len(fitnesses)), key=lambda i: fitnesses[i], reverse=True)
        for i in range(ELITE_SIZE):
            new_population.append(population[sorted_indices[i]][:])

        while len(new_population) < POP_SIZE:
            p1 = selection(population, fitnesses)
            p2 = selection(population, fitnesses)
            c1, c2 = crossover(p1, p2)
            c1 = mutate(c1)
            c2 = mutate(c2)
            new_population.append(c1)
            if len(new_population) < POP_SIZE:
                new_population.append(c2)

        population = new_population

    return best_chromosome, best_profit, best_profit_per_gen

# ============================
# Run BCGA 20 times
# ============================
if __name__ == "__main__":
    num_runs = 20
    base_seed = 42
    results = []
    all_profit_histories = []

    print("Running BCGA with robust repair operator 20 times...\n")

    for run in range(num_runs):
        seed = base_seed + run
        random.seed(seed)
        np.random.seed(seed)

        start_time = time.time()
        try:
            best_sol, best_val, profit_history = genetic_algorithm()
            assign = decode_chromosome(best_sol)
            feasible = is_feasible(assign)
            penalty = 0
            elapsed = time.time() - start_time
            results.append((run+1, best_val, elapsed, feasible, penalty))
            all_profit_histories.append(profit_history)
            print(f"Run {run+1:2d}: Profit = {best_val:.2f}, Time = {elapsed:.2f} s, Feasible = {feasible}")
        except ValueError as e:
            print(f"Run {run+1:2d}: FAILED – {e}")
            # Append dummy data to keep array dimensions consistent
            results.append((run+1, -np.inf, 0, False, 0))
            all_profit_histories.append([-np.inf] * GENERATIONS)

    # Filter out failed runs for summary
    valid_results = [r for r in results if r[1] != -np.inf]
    if not valid_results:
        print("\nAll runs failed – problem is infeasible. Increase capacities or allow unassigned jobs.")
        exit()

    profits = [r[1] for r in valid_results]
    times = [r[2] for r in valid_results]

    avg_profit = np.mean(profits)
    std_profit = np.std(profits)
    best_profit = np.max(profits)
    worst_profit = np.min(profits)
    avg_time = np.mean(times)

    print("\n=== Summary over successful runs ===")
    print(f"Average profit   : {avg_profit:.2f} ± {std_profit:.2f}")
    print(f"Best profit      : {best_profit:.2f}")
    print(f"Worst profit     : {worst_profit:.2f}")
    print(f"Average time     : {avg_time:.2f} s")

    # Plot convergence (skip runs that failed)
    plt.figure(figsize=(10, 6))
    generations = np.arange(1, GENERATIONS + 1)
    for i, history in enumerate(all_profit_histories):
        if history[0] == -np.inf:
            continue
        plt.plot(generations, history, alpha=0.3, linewidth=0.8, label=f'Run {i+1}' if i < 5 else "")

    # Compute average only over successful runs
    valid_histories = [h for h in all_profit_histories if h[0] != -np.inf]
    if valid_histories:
        stacked = np.vstack(valid_histories)
        avg_history = np.mean(stacked, axis=0)
        plt.plot(generations, avg_history, color='black', linewidth=2.5, label='Average Profit')

    plt.xlabel('Generation')
    plt.ylabel('Best Profit (Feasible)')
    plt.title('Profit Convergence of BCGA with Robust Repair')
    plt.legend(loc='lower right')
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()
```
