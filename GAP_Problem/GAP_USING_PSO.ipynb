```python
# using penalty – with monotonically increasing profit curve

import numpy as np
import time
import matplotlib.pyplot as plt

# ====================== PSO Parameters (UNCHANGED) ======================
cost = np.array(cost)
resource = np.array(resource)
capacity = np.array(capacity)
n_agents = 10
n_jobs = 60
swarm_size = 60
max_iter = 300
w_start = 0.9
w_end = 0.4
c1 = 2.0
c2 = 2.0
v_max = 2.0
penalty = 1e6
pos_min = 1.0
pos_max = 11.0 - 1e-6

# ====================== Fitness Function (UNCHANGED) ======================
def evaluate(x):
    a = np.floor(x).astype(int) - 1          # 0‑based agent indices
    profit = np.sum(cost[a, np.arange(n_jobs)])
    used = np.zeros(n_agents)
    for job, agent in enumerate(a):
        used[agent] += resource[agent, job]
    excess = np.sum(np.maximum(0, used - capacity))
    fitness = profit - penalty * excess
    return fitness, a, profit, excess

# ====================== Compute Penalty (UNCHANGED) ======================
def compute_penalty(assign):
    used = np.zeros(n_agents)
    for job, agent in enumerate(assign):
        used[agent] += resource[agent, job]
    excess = np.sum(np.maximum(0, used - capacity))
    return penalty * excess

# ====================== Repair Heuristic (UNCHANGED) ======================
def repair_solution(assign, profit):
    assign = assign.copy()
    used = np.zeros(n_agents)
    for job, agent in enumerate(assign):
        used[agent] += resource[agent, job]
    overloaded = np.where(used - capacity > 0)[0].tolist()
    while overloaded:
        best_delta = -np.inf
        best_move = None
        for i in overloaded:
            jobs_i = [j for j, a in enumerate(assign) if a == i]
            for j in jobs_i:
                for k in range(n_agents):
                    if k == i:
                        continue
                    if used[k] + resource[k, j] <= capacity[k] + 1e-6:
                        delta = cost[k, j] - cost[i, j]
                        if delta > best_delta:
                            best_delta = delta
                            best_move = (j, i, k)
        if best_move is None:
            break
        j, i, k = best_move
        used[i] -= resource[i, j]
        used[k] += resource[k, j]
        assign[j] = k
        profit += best_delta
        overloaded = np.where(used - capacity > 0)[0].tolist()
    return assign, profit

# ====================== Single PSO Run with Monotonic Profit History ======================
def run_pso():
    start = time.time()

    # Initialize
    positions = np.random.uniform(pos_min, pos_max, (swarm_size, n_jobs))
    velocities = np.random.uniform(-1, 1, (swarm_size, n_jobs))

    fitness = np.zeros(swarm_size)
    profits = []
    excesses = []
    for i in range(swarm_size):
        f, a, p, e = evaluate(positions[i])
        fitness[i] = f
        profits.append(p)
        excesses.append(e)

    pbest_pos = positions.copy()
    pbest_fit = fitness.copy()
    pbest_profits = profits.copy()
    pbest_excess = excesses.copy()

    gbest_idx = np.argmax(pbest_fit)
    gbest_pos = pbest_pos[gbest_idx].copy()
    gbest_fit = pbest_fit[gbest_idx]
    gbest_profit = pbest_profits[gbest_idx]
    gbest_excess = pbest_excess[gbest_idx]

    # Track the best profit seen so far (monotonic)
    best_profit_so_far = gbest_profit
    history = []          # store best profit so far at each iteration

    # Main loop
    for t in range(max_iter):
        w = w_start - (w_start - w_end) * (t / max_iter)
        for i in range(swarm_size):
            r1 = np.random.random(n_jobs)
            r2 = np.random.random(n_jobs)
            velocities[i] = (w * velocities[i] +
                             c1 * r1 * (pbest_pos[i] - positions[i]) +
                             c2 * r2 * (gbest_pos - positions[i]))
            velocities[i] = np.clip(velocities[i], -v_max, v_max)
            positions[i] += velocities[i]
            positions[i] = np.clip(positions[i], pos_min, pos_max - 1e-6)

            f, a, p, e = evaluate(positions[i])
            if f > pbest_fit[i]:
                pbest_fit[i] = f
                pbest_pos[i] = positions[i].copy()
                pbest_profits[i] = p
                pbest_excess[i] = e
                if f > gbest_fit:
                    gbest_fit = f
                    gbest_pos = positions[i].copy()
                    gbest_profit = p
                    gbest_excess = e

        # Update best profit seen so far (monotonic)
        if gbest_profit > best_profit_so_far:
            best_profit_so_far = gbest_profit
        history.append(best_profit_so_far)

    # Best found (may be infeasible)
    final_assign = np.floor(gbest_pos).astype(int) - 1
    final_profit = gbest_profit
    final_excess = gbest_excess
    feasible = (final_excess == 0)

    # Repair if infeasible
    if not feasible:
        final_assign, final_profit = repair_solution(final_assign, final_profit)
        feasible = True

    # Final feasibility check
    used_final = np.zeros(n_agents)
    for job, agent in enumerate(final_assign):
        used_final[agent] += resource[agent, job]
    excess_final = np.sum(np.maximum(0, used_final - capacity))
    feasible = (excess_final <= 1e-6)
    final_penalty = penalty * excess_final

    elapsed = time.time() - start
    return final_profit, final_penalty, elapsed, feasible, history

# ====================== Run 20 Times and Collect Histories ======================
n_runs = 20
profits = []
penalties = []
times = []
feasibilities = []
histories = []

print(f"Running PSO {n_runs} times...\n")
for run in range(1, n_runs + 1):
    profit, penalty_val, elapsed, feasible, hist = run_pso()
    profits.append(profit)
    penalties.append(penalty_val)
    times.append(elapsed)
    feasibilities.append(feasible)
    histories.append(hist)
    print(f"Run {run:2d}: Profit = {profit:.2f}, Penalty = {penalty_val:.2f}, Time = {elapsed:.2f} s, Feasible = {feasible}")

# Statistics
profits = np.array(profits)
penalties = np.array(penalties)
times = np.array(times)
print("\n=== Summary over 20 runs ===")
print(f"Average profit   : {np.mean(profits):.2f} ± {np.std(profits):.2f}")
print(f"Best profit      : {np.max(profits):.2f}")
print(f"Worst profit     : {np.min(profits):.2f}")
print(f"Average time     : {np.mean(times):.2f} s")
print(f"Total time       : {np.sum(times):.2f} s")
print(f"All runs feasible: {all(feasibilities)}")
print(f"All final penalties zero? {all(p == 0 for p in penalties)}")

# ====================== Plot Monotonically Increasing Profit Convergence ======================
plt.figure(figsize=(10, 6))

# Plot each run's history (now monotonic) with low opacity
for i, hist in enumerate(histories):
    plt.plot(range(1, max_iter+1), hist, alpha=0.3, color='blue', linewidth=0.8)

# Compute and plot average profit history
avg_history = np.mean(histories, axis=0)
plt.plot(range(1, max_iter+1), avg_history, color='red', linewidth=2, label='Average Profit')

plt.xlabel('Iteration')
plt.ylabel('Best Profit (monotonically increasing)')
plt.title('Convergence of PSO over 20 runs (monotonic profit)')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
